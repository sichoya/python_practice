```python
def copy_image(source_file, destination_file):
    """이미지 파일을 복사합니다."""

    try:
        # 원본 파일 읽기
        with open(source_file, 'rb') as source:
            image_data = source.read()

        # 원본 파일 쓰기
        with open(destination_file, 'wb') as destination:
            destination.write(image_data)


        print(f"{source_file}가{destination_file}로 복사 되었습니다.")
        return True
    
    except FileNotFoundError:
        print("Error")
        return False
    except Exception as e:
        print(f"Error {e}")
        return False


copy_image('new.jpg', 'copy.jpg')
```

### XOR 암호화를 사용한 파일암호화/복호화
<!-- - 과제로 생각하심 추후 학습에 추가할 것.
 -->


```python
def XOR_encrypt_decrypt(input_file, output_file, key):
    """암호화/복호화"""
    try:
        # 원본 파일 읽기
        with open(input_file, 'rb') as infile:
            data = infile.read()

        # 원본 파일 쓰기
        key_bytes = key.encode() if isinstance(key, str) else bytes([key])
        key_len = len(key_bytes)

        encrypted_data = bytearray(len(data))
        for i in range(data):
            encrypted_data[i] = data[i] ^ key_bytes[i & key_len]
    

        with open(output_file, 'wb') as outfile:
            output_file.write(encrypted_data)

        print(f'파일 처리 완료: {input_file} -> {output_file}')
        return True
    
    except Exception as e:
        print(f"오류 발생 {e}")
        return False
    

# 암호화
XOR_encrypt_decrypt('example.txt', 'secret.enc', 'mykey123') # 기존의 파일 있을 시. ㅏ능

# 복호화 (그대로)
# XOR_encrypt_decrypt('example.txt', 'secret.enc', 'mykey123')

```
